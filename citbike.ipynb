```python
import pandas as pd
from pathlib import Path
import datetime as dt
import numpy as np
```

```python

#Read in csv files into pandas dataframes
df_2023 = pd.read_csv(Path('may_2023.csv'), parse_dates=True, infer_datetime_format=True, low_memory=False)

```

```python
#print column names
df_2023.columns
```

```python
#drop ride_id and rideable_type 
df_2023 = df_2023.drop(['ride_id', 'rideable_type'], axis=1)
df_2023.head()
```

```python
start_date_2023 = pd.to_datetime(df_2023['started_at'], format = '%Y-%m-%d %H:%M:%S')
stop_date_2023 = pd.to_datetime(df_2023['ended_at'], format = '%Y-%m-%d %H:%M:%S')
df_2023['started_at'] = start_date_2023
df_2023['ended_at'] = stop_date_2023
month_start = df_2023['started_at'].dt.strftime('%b')
month_stop = df_2023['ended_at'].dt.strftime('%b')
df_2023['start_month'] = month_start
df_2023['stop_month'] = month_stop
df_2023['time_usage'] = df_2023['ended_at'] - df_2023['started_at']
df_2023.head()
```

```python
seconds = df_2023['time_usage'].dt.total_seconds()
df_2023['time_usage_seconds'] = seconds
df_2023.head()
```

```python
df_2023.drop(columns=['stop_month'], inplace=True)
df_2023.head()
```

```python
df_2023.rename(columns={'start_month':'May'}, inplace=True)
df_2023.head()
```

```python
day = df_2023['started_at'].dt.strftime('%a')
df_2023['day'] = day
df_2023.head()
```

```python
start_lat = df_2023['start_lat']
start_lng = df_2023['start_lng']
end_lat = df_2023['end_lat']
end_lng = df_2023['end_lng']
end_lng.head()
```

```python
def haversine_vectorize(start_lat, start_lng, end_lat, end_lng, to_radians=True, earth_radius=6371):
    if to_radians:
        start_lat, start_lng, end_lat, end_lng = np.radians([start_lat, start_lng, end_lat, end_lng])
        a = np.sin((end_lat - start_lat)/2.0)**2 + np.cos(start_lat) * np.cos(end_lat) * np.sin((end_lng - start_lng)/2.0)**2
        return earth_radius * 2 * np.arcsin(np.sqrt(a))
df_2023['distance'] = haversine_vectorize(start_lat, start_lng, end_lat, end_lng)
```
